```
# Ensamblador especifico y Musa:
def ensambladorMini(frase:str, clave:str, insercion:list) -> str:
  '''Metodo que busca una palabra especifica en una frase y la remplaza por
     las palabras que se le pasen en lista
     Parametros:
         frase (str): una frase la cual editar
         clave (str): una palabra que va a buscar en la frase, solo si esta entre parentesis cuadrados
         insercion (list(str)): es una lista de palabras que se inserta donde este la clave
     Retorno:
         retorna la frase con las palabras insertadas
     Ejemplo:
         ensambladorMini("[personaje] and [personaje] in [lugar]","personaje",["Jose","Juliet"]):
             retorna -> "Jose and Juliet in [lugar]"
         El metodo solo inserta las palabras en donde encuentra la clave y remueve solo los corchetes 
         que tienen la clave'''
  if insercion == None or insercion == []:
    return frase # si no hay nada que insertar
  fin = str("#" *random.randint(1, 5) + "%" * random.randint(1, 5) + "$" * random.randint(1, 5))
  frase = " " + frase + fin  # esto es para evitar los casos en que la frase comience o termine con la clave
  j = 0
  
  for i in frase.split("[" + clave +"]"):
    if j == 0:
      frase = i # esto sobreescribe la frase, para no crear una variable
    else:
      frase += i
    if fin not in i:
      try: frase += insercion[j] # esto es para los casos en que la lista trae mÃ¡s inserciones de las que hay en la frase
      except:
        print("Una excepcion a ocurrido",frase,
              "necesitaba mÃ¡s incersiones de las que le fueron pasadas en el",
              "parametro insercion")
        frase += "[" + clave + "]"
    j += 1
  return frase.split(fin)[0][1:] #esto retorna la frase


# Ensamblador de frases y Ensayos
#(este debe ser modificado para utilizar las claves de multiplicidad y use un diccionario)
def ensambladorFrase(frase:str, personajes:list, lugares:list, tiempos:list, acciones:list, eventos:list) -> str:
    '''Metodo que recibe una frase con todos los ensambladores posibles
       y devulve la frase ensamblada
       Parametros:
          frase (str): una frase la cual editar que puede o no contener
                       espacio para insertar el resto de parametros
          personajes (list(str)): una lista de palabras a insertar donde
                                  encuentre la palabra personaje
          lugares (list(str)): una lista de palabras a insertar donde
                               encuentre la palabra lugar
          tiempos (list(str)): una lista de palabras a insertar donde
                               encuentre la palabra tiempo
          acciones (list(str)): una lista de palabras a insertar donde
                                encuentre la palabra accion
          eventos (list(str)): una lista de palabras a insertar donde
                               encuentre la palabra evento
       Retorno:
          retorna la frase con las palabras insertadas
       Ejemplo:
          ensambladorMini("[personaje] and [personaje] in [lugar]",["Jose","Juliet"], ["Kingdom"],["yesterday"],["fly"],["pee"]):
              retorna -> "Jose and Juliet in Kingdom"
          El metodo inserta las palabras que encuentre en la frase, si no
          encuentra una clave, aunque se le haya pasado una lista, no las inserta'''

    return ensambladorMini(
        ensambladorMini(
            ensambladorMini(
                ensambladorMini(
                    ensambladorMini(frase, "personaje", personajes),
                          "lugar", lugares),
                    "tiempo", tiempos),
                "accion", acciones),
            "evento", eventos) #retorna usando el metodo ensambladorMini(), debe modificarse
```

###ExtracciÃ³n y segregaciÃ³n de contenido

```
# Retorno aleatorio y Ordenanza
def obtenerAleatorio(L:list):
  '''Metodo que retorna un valor aleatorio en una lista enumerada
       Parametros:
          L (list): una lista
       Retorno:
          retorna un valor cualquiera que contenga L
       Ejemplo:
          obtenerAleatorio(["Jose","Juliet","Rosero","Fabian"]):
             retorna -> "Jose"'''
  while True:
    i = random.randint(0,len(L)-1)
    if L[i]: return L[i]
    

def desordenar(L:list)->list:
  '''metodo que recibe un list y lo devuelve en un orden aleatorio
      Parametros:
          L (list): una lista
      Retorno:
          retorna L con sus elementos intercambiados aleatoriamente
      Ejemplo:
          desordenar(["Jose","Juliet","Rosero","Fabian"]):
             retorna -> ['Fabian', 'Rosero', 'Jose', 'Juliet']'''
  try:
    orden = [] # aqui se almacenan los nuevos indices
    for i in L:
      while True:
        rand = random.randint(0,len(L)-1) # genera un numero aleatorio en el rango de L
        if rand not in orden:
          orden.append(rand)
          break
    return [L[i] for i in orden] # retorna L en el nuevo orden de indices
  except: return L # en caso de error retorna nuevamente L
```

```
# Elimina lo inconcluso en una frase y Filtro EMI
def extraerFrase(texto:str) -> str:
  '''metodo que toma el ultimo punto autoconclusivo de un texto y retorna todo lo anterior a este
    Parametros:
        texto (str): un fragmento de texto
    Retorno:
         retorna frase despues de haberle quitado frases inconclusas
     Ejemplo:
        extraerFrase("Fabian is a Great Player. The first thing that stands out 
                      about him, and I think it's important to note here because
                      he has been so good for the team this year (and in his 
                      rookie season), but also what makes Adrian an exceptional 
                      player. He doesn't just play well on defense; when you 
                      watch video of games where they're playing against each 
                      other or")
        retornara -> "Fabian is a Great Player. The first thing that stands out 
                      about him, and I think it's important to note here because
                      he has been so good for the team this year (and in his 
                      rookie season), but also what makes Adrian an exceptional 
                      player. He doesn't just play well on defense;'
      Como se puede ver el metodo debe eliminar la ultima frase, pues el modelo
      no pudo concluirla por nuestro limite de tamaÃ±o. Notese que no siempre es 
      a partir de un punto lo que se busca eliminar, pues en el ejemplo es un ";",
      pero en escecia siempre debe ser una puntuacion el final.
      Ademas, no importa si queda ";", "," o "." al final, el retorno sea con un
      punto al final'''
  import nltk
  nltk.download('punkt')
  if texto.replace(nltk.sent_tokenize(texto)[-1], ""):
    return texto.replace(nltk.sent_tokenize(texto)[-1], "")
  return texto + ".\n"
```

```
# Identifica las categorias en una frase y la descompone y Pistas
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
def extraerCategorias(frase:str) -> dict:
  '''metodo que retorna un diccionario con las categorias identificadas en la frase mÃ¡s la misma
      Parametros:
        frase (str): una frase en la cual se basara el metodo
                     para generar el fragmento
      Retorno:
         Diccionario con las categorias identificadas y la frase original
      Ejemplo:
        extraerCategorias("so Groucho say, \"I'm not a man of my word.\".\nThe first time I saw 
                           the film was in New York City at an event called The Film Society. It
                           had been screened there before and it seemed to be well received by everyone")
          retornara -> {'frase': 'so Groucho say, "I\'m not a man of my word.".\nThe first time
                                  I saw the film was in New York City at an event called The 
                                  Film Society. It had been screened there before and it seemed
                                  to be well received by everyone',
                        'personajes': ["Groucho","everyone"],
                        'lugares': ["New York City"],
                        'tiempos': ["The first time","before"]},
                        'acciones': ["say","saw","was","had been","seemed","received"],
                        'eventos': ["saw the film","The Film Society","to be well received"] '''
  import nltk
  nltk.download('punkt')
  nltk.download('averaged_perceptron_tagger')

  tagged = nltk.pos_tag(nltk.word_tokenize(frase))
  personajes = []
  lugares = []
  eventos = []
  i = 0
  while i < len(tagged):
    if tagged[i][1] == "NNP":
      ins = personajes
      if i > 0 and tagged[i-1][1] == "IN":
        ins = lugares
      ins.append(tagged[i][0])
      for j in range(i + 1, len(tagged)):
        if tagged[j][1] != "NNP":
          i = j
          break
        ins[-1] += " " + tagged[j][0]
    if tagged[i][1] == "NN":
      for j in range(i - 5, i):
        try:
          if tagged[j][1][:2] == "VB" and tagged[j][0].isalpha():
            eventos.append(tagged[j][0])
            for k,q in tagged[j + 1: i + 1]:
              eventos[-1] += " " + k
            break
        except: continue
    i += 1
  return {"frase": frase, "personajes": personajes, "lugares": lugares,
          "tiempos": [i.lower() for i, j in tagged if j[:2] == "RB"],
          "acciones": [i for i, j in tagged if j[:2] == "VB" and i.isalpha()],
          "eventos": eventos}
```

###Depurado de historias

```
# Retorno y Tecnico
#(este debe filtrar las frases, para no repetir y esas vainas)
def depuradorHistoria(frase:str):
  '''metodo que retorna un fragmento de historia
    Parametros:
        frase (str): un fragmento de texto
    Retorno:
         retorna frase despues de haberle quitado incoherencias
     Ejemplo:
         depuradorHistoria("finally invasion of the United States.
                         the first step in this process was to establish a 
                         new government, which would be called "the Republic." 
                         The Constitution"):
             retorna -> "Finally invasion of the United States.
                         The first step in this process was to establish a 
                         new government, which would be called "the Republic"."
         El metodo convirtio en mayuscula los inicio de los parrafos y omitio
         la frase que estaba comenzando'''
  frase = extraerFrase(frase) # extrae las frases inconclusas utilizando el metodo extraerFrase()
  frase = frase[0].upper() + frase[1:] # convierte a mayuscula el inicio de la entrada
  return frase
```

```
# Adapta una frase con referencias a particularidades de la historia y Cover
def adapter(frase:str,referencias:list,especificacion:list)->str:
  '''metodo que toma una frase que construimos a punta de referencias y debe eliminar todas las referencias y reemplazarlas por otras palabra
  ejemplo de como deberia quedar:
  adapter("Rambo in The White House then, and now.",["Rambo","The White House","then"],["Jose","his township","at quarantine"])
    retornara -> 'Jose in his township at quarantine, and now' '''
  if especificacion == [] or not especificacion or None in especificacion or referencias  == []:
    return frase
  else:
    return adapter(frase.replace(referencias[0], especificacion[0]), referencias[1:], especificacion[1:])
```

```
def obtenerIndex(direccion:dict = None, super = CATEGORIAS):
  if type(super) == dict and super != {}:
    if type(direccion) == dict and direccion != {}:
      if list(direccion.keys())[0] in super:
        return obtenerIndex(direccion[list(direccion.keys())[0]], super[list(direccion.keys())[0]])
    if direccion:
      try: return super[direccion]
      except:
        try: return super[list(super.keys())[direccion.keys()]]
        except: print("Error de llave")
  if type(super) == list and super != []:
    if type(direccion) == int:
      try: return super[direccion]
      except IndexError: print("Error de index")
    if type(direccion) == tuple and len(direccion) == 2:
      try: return [super[i] for i in range(direccion[0],direccion[1])]
      except IndexError: print("Error de index")
  print("Error en direccion:", direccion, "para", super)

# Generador de especificaciones y Peticion del Publico
#(este debe ser modificado para utilizar dinamicamente las frases)
def peticionFrase(especificaciones:dict = None) -> str:
  '''metodo que permite generar frases a traves de multiples parametros, 
     esto se hace solo para que las pruebas sean mÃ¡s rapidas, para saber 
     que parametros se le deben pasar a ensambladorMini() o a ensambladorFrase()
     
     Parametros:
         especificaciones (dict): Diccionario que especifica que claves usar
                                  para generar la frase
      Retorno:
         La frase con las referencias cambiadas segun las espedificaciones
      Ejemplo:
      1 -> peticionFrase({"frase":{"claves":{"NN":0}}})
            retornara -> 'Zelda is a'
          En este ejemplo el metodo revisa el diccionario por parametro y solo 
          ve que contiene la llave "frase", "claves", acude al diccionario 
          CATEGORIAS para ver si existe ese diccionario, como sÃ­ lo encuentra 
          entonces ahora sabe que la frase que debe retornar esta en el 
          diccionario CLAVES, sigue llendo mÃ¡s adentro y ve la llave "NN", la 
          busca en el diccionario CLAVES y halla el numero 0, busca si 
          CLAVES["NN"] tiene ese index, al hallarlo toma eso como frase. 
          Como el parametro al llamar el metodos solo traia la llave "frase" 
          entonces el decide llenar esa frase con un genero aleatorio, en el 
          ejemplo, generos debio quedar "Aventura", por eso termino siendo Zelda.

      2 -> peticionFrase({"frase":{"claves":{"NN":0}}, "genero":"Fantasia"})
            retornara -> 'Link is a'
          En este ejemplo se sigue la logica del anterio, asi el metodo sabe
          que la frase que debe devolver es '[personaje] is a', ahora, como
          en el parametro del metodo tambien encuentra que se especifico el
          genero "fantasia".

      3 -> peticionFrase({"frase":{"claves":{"NN":0}},"personajes":["Mario","Luigi"]})
            retornara -> 'Mario is a'
          En este caso, siguiendo la logica del primer ejemplo, la frase que
          debe devolver es '[personaje] is a', ahora, como en el parametro no
          especifica genero, el metodo escojera uno aleatorio, pero sÃ­ esta las 
          claves "personajes" entonces se busca dentro de esa llave y halla un 
          list (["Mario","Luigi"]) entonces el retorno serÃ¡ GENERO = obtenerAleatorio(GENEROS);
      return ensambladorFrase(CATEGORIAS["claves"]["NN"][0], especificaciones["personajes"],
                              LUGARES[GENERO], TIEMPOS[GENERO], ACCIONES[GENERO],
                              EVENTOS[GENERO])
          Notese que si en "personajes" hubiera hallado un dict entonces 
          tendria que buscar las llaves de ese dict, y si eran,
          por ejemplo {"personajes":{"Fantasia":(0,2)}} entonces el 
          retorno hubiese sido GENERO = obtenerAleatorio(GENEROS);
      return ensambladorFrase(CATEGORIAS["claves"]["NN"][0], 
                              [CATEGORIAS["personajes"]["Fantasia"][i] for i in range(0,2)],
                              LUGARES[GENERO], TIEMPOS[GENERO], ACCIONES[GENERO], EVENTOS[GENERO])
      
      4 -> peticionFrase()
            retornara -> 'Rick Sanchez in spaceship then pandemic'
          En este caso como no recibe ningun parametro, el toma una frase
          aleatoria de CLAVES["NN" entonces el retorno serÃ¡ GENERO = obtenerAleatorio(GENEROS);
      return ensambladorFrase(obtenerAleatorio(CATEGORIAS["claves"]["NN"]),
                              PERSONAJES[GENERO], LUGARES[GENERO], TIEMPOS[GENERO],
                              ACCIONES[GENERO], EVENTOS[GENERO])
          
      NOTAS:
      - si en el diccionario parametro, la llave "frase" es un str, entonces
        directamente lo tomara como frase, asumiendo que es una frase con
        categorias a ensamblar
      - si en el diccionario parametro, la llave "genero" es un int, entonces
        tomara el genero CATEGORIAS["generos"][int], siendo int obviamente el
        entero las claves que se deben buscar de momento en el parametro son:
        "frase", "genero", "personajes", "lugares", "tiempos", "eventos" y 
        "acciones", cualquier otra llave sera ignorada de momento
      - si en algun punto recibe un tuple de int (como en la segunda parte del
        ejemplo 3), entonces el metodo buscara si la lista solicitada tiene 
        todos los indices en este rango, si los tiene entonces generara un list
        con estos indices
      - si el metodo va a buscar una llave o un indice en un diccionario o una
        lista y no la encuentra, debe retornarse a si mismo sin parametros, 
        asi: return peticionFrase()
      - si el metodo recibe cualquier otra cosa que no sea un dict debe 
        retornarse a si mismo sin parametros, asi: return peticionFrase()
      - si el metodo recibe la llave "generos", y en esta no hay un int o un 
        str debe retornarse a si mismo sin parametros, asi: return peticionFrase()
      - si en algun diccionario encuentra un str, pero deberia encontrar un 
        list o un dict, debe retornarse a si mismo sin parametros, asi: 
        return peticionFrase()
        ejemplo: si recibe peticionFrase({"frase":{"claves":{"NN":0}}, "genero":"Aventura", 
                                            "personajes":"Mario"}),
        pues en "personajes" solo debe hallar un list o un dict '''
  frase = None
  personaje = None
  genero = None
  tiempo = None
  lugar = None
  accion = None
  evento = None
  
  if especificaciones and type(especificaciones) == dict and especificaciones is not {}:
    if "frase" in especificaciones:
      if type(especificaciones["frase"]) == str:
        frase = especificaciones["frase"]
      if type(especificaciones["frase"]) == int:
        try: frase = CATEGORIAS["claves"]["NN"][especificaciones["frase"]]
        except: pass
      if type(especificaciones["frase"]) == dict:
        frase = obtenerIndex(especificaciones["frase"])
            
    if 'genero' in especificaciones :
      if type(especificaciones['genero']) == str:
        genero = especificaciones['genero']
      if type(especificaciones['genero']) == int:
        try: genero = CATEGORIAS['generos'][especificaciones['genero']]
        except: pass
      if type(especificaciones["genero"]) == dict:
        genero = obtenerIndex(especificaciones["genero"])

    if ('personaje' in especificaciones):
      if type(especificaciones['personaje']) == list:
        personaje = especificaciones['personaje']
      if type(especificaciones['personaje']) == dict:
        personaje = obtenerIndex(especificaciones["personaje"])
      if type(especificaciones['personaje']) == str:
        personaje = especificaciones['personaje']
      if genero and (type(especificaciones['personaje']) == tuple or
                     type(especificaciones['personaje']) == int):
        personaje = obtenerIndex(especificaciones["personaje"], super = CATEGORIAS["personajes"][genero])

    if ('lugar' in especificaciones):
      if type(especificaciones['lugar']) == list:
        lugar = especificaciones['lugar']
      if type(especificaciones['lugar']) == dict:
        lugar = obtenerIndex(especificaciones["lugar"])
      if type(especificaciones['lugar']) == str:
        lugar = especificaciones['lugar']
      if genero and (type(especificaciones['lugar']) == tuple or
                     type(especificaciones['lugar']) == int):
        lugar = obtenerIndex(especificaciones["lugar"], super = CATEGORIAS["lugares"][genero])

    if ('tiempo' in especificaciones):
      if type(especificaciones['tiempo']) == list:
        tiempo = especificaciones['tiempo']
      if type(especificaciones['tiempo']) == dict:
        tiempo = obtenerIndex(especificaciones["tiempo"])
      if type(especificaciones['tiempo']) == str:
        tiempo = especificaciones['tiempo']
      if genero and (type(especificaciones['tiempo']) == tuple or
                     type(especificaciones['tiempo']) == int):
        tiempo = obtenerIndex(especificaciones["tiempo"], super = CATEGORIAS["tiempos"][genero])

    if ('accion' in especificaciones):
      if type(especificaciones['accion']) == list:
        accion = especificaciones['accion']
      if type(especificaciones['accion']) == dict:
        accion = obtenerIndex(especificaciones["accion"])
      if type(especificaciones['accion']) == str:
        accion = especificaciones['accion']
      if genero and (type(especificaciones['accion']) == tuple or
                     type(especificaciones['accion']) == int):
        accion = obtenerIndex(especificaciones["accion"], super = CATEGORIAS["acciones"][genero])

    if ('evento' in especificaciones):
      if type(especificaciones['evento']) == list:
        evento = especificaciones['evento']
      if type(especificaciones['evento']) == dict:
        evento = obtenerIndex(especificaciones["evento"])
      if type(especificaciones['evento']) == str:
        evento = especificaciones['evento']
      if genero and (type(especificaciones['evento']) == tuple or
                     type(especificaciones['evento']) == int):
        evento = obtenerIndex(especificaciones["evento"], super = CATEGORIAS["eventos"][genero])

  if not genero or type(genero) != str: genero = obtenerAleatorio(GENEROS)
  if not frase or type(frase) != str: frase = obtenerAleatorio(CLAVES["NN"] + CLAVES[genero])
  if not personaje: personaje = obtenerAleatorio(PERSONAJES[genero])
  if not lugar: lugar = obtenerAleatorio(LUGARES["NN"] + LUGARES[genero])
  if not accion: accion = obtenerAleatorio(ACCIONES["NN"] + ACCIONES[genero])
  if not evento: evento = obtenerAleatorio(EVENTOS["NN"] + EVENTOS[genero])
  if not tiempo: tiempo = obtenerAleatorio(TIEMPOS["NN"] + TIEMPOS[genero])

  if type(personaje) != list: personaje = [personaje]
  if type(lugar) != list: lugar = [lugar]
  if type(tiempo) != list: tiempo = [tiempo]
  if type(accion) != list: accion = [accion]
  if type(evento) != list: evento = [evento]

  frase = ensambladorFrase(frase,personaje,lugar,tiempo,accion,evento) # con todos los parametros construye una frase
  
  return frase
```

```
# Codigo de control y Discografica
class Historia:
  id = None
  titulo = None
  genero = None
  personajes = None
  lugares = None
  tiempos = None
  eventos = None
  acciones = None
 
  def __init__(self,genero:str, titulo:str = None, fichas_tecnicas:dict = {}):
    self.id=makeId()
    IDs[self.id] = self
    self.genero = genero
    for i in ["personajes", "lugares", "tiempos", "eventos", "acciones"]:
      CATEGORIAS[self.id + str(i)] = []
    for i,j in fichas_tecnicas.items():
      agregarACategoria({self.id + i:j})
    self.personajes = CATEGORIAS[self.id + "personajes"]
    self.lugares = CATEGORIAS[self.id + "lugares"]
    self.tiempos = CATEGORIAS[self.id + "tiempos"]
    self.eventos = CATEGORIAS[self.id + "eventos"]
    self.acciones = CATEGORIAS[self.id + "acciones"]
 
  def agregar(self, dicCate):
    for i,j in dicCate.items():
      agregarACategoria({self.id+i:j})
    self.personajes += dicCate["personajes"]
    self.lugares += dicCate["lugares"]
    self.tiempos += dicCate["tiempos"]
    self.eventos += dicCate["eventos"]
    self.acciones += dicCate["acciones"]
  
  def __referencias(self, nuevo:bool = True):
    '''Metodo que retorna las referencias para los personajes, lugares, tiempos,
        acciones, eventos y la clave de las frases
        Parametros:
          nuevo (bool): True-> Utiliza nuevas referencias para la creaciÃ³n de la frase
                        False -> Toma las referencias existentes y le agrega mÃ¡s
        Retorno:
          Una lista de las referencias en el siguiente orden:
            Personajes, Lugares, Tiempos, Accikones, Eventos, Clave'''
    if nuevo:
      refP = desordenar(PERSONAJES[self.genero])
      refL = desordenar(LUGARES[self.genero])
      refT = desordenar(TIEMPOS[self.genero])
      refA = desordenar(ACCIONES[self.genero])
      refE = desordenar(EVENTOS[self.genero])
    else:
      refP = desordenar(PERSONAJES[self.genero] + self.personajes)
      refL = desordenar(LUGARES[self.genero] + self.lugares)
      refT = desordenar(TIEMPOS[self.genero] + self.tiempos)
      refA = desordenar(ACCIONES[self.genero] + self.acciones)
      refE = desordenar(EVENTOS[self.genero] + self.eventos)
    ref_clave = obtenerAleatorio(desordenar(CLAVES["NN"])) # En el caso de que CLAVES tenga mÃ¡s datos para cada genero se debe cambiar a CLAVES[self.genero]
    return refP, refL, refT, refA, refE, ref_clave

  def __crear_frase(self, refP:list, refL:list, refT:list, refA:list, refE:list,
                      ref_clave:str, finalizar:bool = False):
    '''Metodo que crea una frase tomando los artibutos como las referencias
         Parametros:
          refP (List): Lista con las referencias de Personajes
          refL (List): Lista con las referencias de Lugares
          refT (List): Lista con las referencias de Tiempos
          refA (List): Lista con las referencias de Acciones
          refE (List): Lista con las referencias de Eventos          
          ref_clave (str): Estructura de la frase a crear
          finalizar (bool)[opcional]: True -> Forza una frase de finalizaciÃ³n
                                      False -> Crea una frase con una clave aleatorioa
        Retorno:
          La frase generada por GTP-2 a partir de los parametros dle metodo'''
    if finalizar:
      ensamblaF = peticionFrase({"frase": CLAVES["NN"][(3, 4, 5, 11, 12, 17)[random.randint(0, 2)]],
                                 "genero":self.genero, "personajes": refP, "lugares": refL, 
                                 "tiempos": refT, "acciones": refA,"eventos": refE})
    else:
      ensamblaF = ensambladorFrase(ref_clave, refP, refL, refT, refA, refE)
    generaH = generadorHistoria(ensamblaF)
    frase = adapter(generaH, refP + refL + refT + refA +refE, self.personajes + self.lugares +
                      self.tiempos + self.acciones + self.eventos)
    return frase

  # Extraer las partes de la historia 
  def iniciarHistoria(self) -> str:
    '''Metodo que crea una historia aleatoria y la adapta a las configuraciones
       de la clase Historia
       Parametros:
          El metodo no tiene parametros
       Retorno:
          Frase generada con GTP-2 y las categorias de la frase
       Ejemplo:
          obj_historia = Historia("Aventura")
          obj_historia.iniciarHistoria()
          Esto retornara una frase del genero Aventura con los personajes, lugares,
          tiempos, eventos y acciones correspondientes al objeto obj_historia'''
    # Definicion de referencias aleatorias
    refP, refL, refT, refA, refE, ref_clave = self.__referencias()
    
    # GeneraciÃ³n de la frase a partir de las referencias y modificaciÃ³n por los atributos de la clase
    retorno = self.__crear_frase(refP, refL, refT, refA, refE, ref_clave)
    
    # IdentificaciÃ³n de cada una de las categorias que hay en la frase
    categorias = extraerCategorias(retorno)
    del categorias['frase']
    self.agregar(categorias)
    return retorno
 
  # Generear la hisotira
  def continuarHistoria(self) -> str:
    # Definicion de referencias aleatorias
    refP, refL, refT, refA, refE, ref_clave = self.__referencias(False)
    
    # GeneraciÃ³n de la frase a partir de las referencias y modificaciÃ³n por los atributos de la clase
    retorno = self.__crear_frase(refP, refL, refT, refA, refE, ref_clave)
    
    # IdentificaciÃ³n de cada una de las categorias que hay en la frase
    categorias = extraerCategorias(retorno)
    del categorias["frase"]
    
    # Control de la cantidad de los multiples llamos a la generaciÃ³n de una historia
    if categorias["eventos"] and categorias["eventos"] != []: # Valida que se hayan generado eventos
        exponEventos = len(categorias["eventos"])
        self.agregar(categorias)
        if len(self.eventos) - exponEventos > exponEventos: # Valida que la historia sea concluyente
          return retorno
        else:
          if len(self.eventos) > (exponEventos**2): # Forzar una historia concluyente
            retorno = self.__crear_frase(refP, refL, refT, refA, refE, ref_clave, True)
          return retorno + self.continuarHistoria()
    else:
      print("Â¡No produjo eventos!")
      return self.continuarHistoria()
 
  def controlador(self):
      return
 
  def getFichaTecnica(self):
    return {i[len(self.id):]:j for i,j in CATEGORIAS.items() if self.id in i}
```
